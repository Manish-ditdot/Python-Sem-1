# Now lets generate a hundred sequences, each of which is 5 bits larger than the previous one. 

```python
#problem 2

n=1000
def prob(n):
    p=[]
    for i in range(0,n):
        x=np.random.rand(100+i*5)
        one=[]
        zero=[]
        for j in range(0,len(x)):
            if x[j]<=(1/3):
                zero.append(0)
            else:
                one.append(1)
        p0=len(zero)/(len(zero)+len(one))
        p.append(p0)
    return p
plt.plot(prob(n))
```

## The relationship of frequencies to underlying probability distributions is at the core of what is called the "frequentist vs Bayesian" view of statistics.

```python
#problem 3

n=1000
def prob(n):
    p=[]
    for i in range(0,n):
        x=np.random.rand(100+i*5)
        one=[]
        zero=[]
        for j in range(0,len(x)):
            if x[j]<=(1/3):
                zero.append(0)
            else:
                one.append(1)
        p0=len(zero)/(len(zero)+len(one))
        p.append(p0)
    return p
plt.figure(figsize=(16,10))
for i in range(20):
    
    plt.plot(prob(n))
```

# What we are seeing here is the saturation of the probability of the zeros (and hence the ones) to the PDF. This can be captured via the notion of a sequence entropy.

# $P[x_1,x_2\ldots x_n]=P_{X}(a_1)^{nP(a_1)}\ldots P_{X}(a_k)^{n P(a_k)}$

# $H_{seq}(x)=-\frac{1}{n}\log(P_{X^n}(x^n))$ is the sequence entropy. It converges to the binary entropy.
